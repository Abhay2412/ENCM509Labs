## Introduction 
For Exercise 3, the dataset appears to represent the demographic data of passengers and crew aboard the Diamond Princess cruise ship during the COVID-19 pandemic in February 2020. This data can be used to establish biometric enabled decisions for understanding the distribution of age and gender which can be crucial for the cruise ship industry in managing health and safety protocols. The dataset includes the percentage of male and female individuals as the count of people within specific age intervals. 

### Bar Chart
We will plot the bar chart first similar to the figure 2 in the lab instructions. With the age intervals on the x-axis and the quantity of people on the y-axis.

````python
# Data provided in the exercise regarding the gender, age and age labels 
gender = [55, 45] # 55% of male and 45% of female

# counting of people at each age interval shown in age labels
age = [16, 23, 347, 428, 334, 398, 923, 1015 , 216, 11]
age_labels = ['0-9','10-19','20-29','30-39', '40-49','50-59','60-69','70-79','80-89','90-99']
````

````python
# Importing the library for making plots 
import matplotlib.pyplot as plt

# Set the figure size of the bar chart
plt.figure(figsize=(10, 6))

# The configuration of the bar chart
plt.bar(age_labels, age, color='blue', label='Age distribution')

# Add data points and connect them with a trend line
for i, v in enumerate(age):
    plt.text(i, v + 25, str(v), ha='center', color='blue')
plt.plot(age_labels, age, color='red', marker='o', linestyle='--', linewidth=2, markersize=5, label='Trend line')

# Setting the title and the x/y-axis labels
plt.title('Age Distribution of a Cruise Ship during COVID-19')
plt.xlabel('Age Groups')
plt.ylabel('Number of People')
plt.legend()
````

## Analysis of the Bar Chart
The bar chart above displays the age distribution of the individuals aboard a cruise ship during the COVID-19 outbreak. The age groups which are plotted on the x-axis and the number of the people in each group is represented on the y-axis. From the chart it is evident that the age groups of 70-79 years old and 60-69 years old have the highest number of people. This is followed by the age groups of 50-59 years old and 40-49 years old, which show substantial numbers as well. The age groups of 0-9 years old, 10-19 years old and 90-99 years old have the lowest number of people indicating that there were relatively few children and elderly people aboard the ship. The trend line depicted by the dashed red line highlights the concentration of the individuals in the 40-79 years old age groups. This concentration of older individuals could have could have significant implications for health and safety measures as older age groups are more susceptible to COVID-19. Overall this chart provides valuable insight into the demographic distribution of the ship's population. 

### Pie Chart
We will plot the pie chart similar to the figure 2 in the lab instructions. Which will show the distribution of the male and female individuals aboard the ship.

````python
# Plotting the pie chart for gender distribution
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.pie(gender, labels=['Male', 'Female'], autopct='%1.1f%%')
plt.title('Gender Distribution')
````

## Analysis of the Pie Chart
The pie chart illustrates the gender distribution of passengers and crew on a cruise ship, it indicates a relatively balanced distribution with a slight majority of males at 55% and females at 45%. This near parity in gender distribution is important for designing and implementing health and safety protocols, as it suggests that any measures taken should be universally applicable to all individuals regardless of gender. Additionally, in the context of the COVID-19 pandemic, understanding the gender distribution is crucial since research has shown that the virus can affect genders differently. Overall, the gender distribution data is a key demographic factor that aids in the comprehensive assessment of the ship's population, ensuring that decision support systems can be effectively tailored to the needs of all individuals.
