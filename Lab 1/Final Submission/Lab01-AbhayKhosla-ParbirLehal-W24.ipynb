## Introduction 


In Exercise 4, we delve into the statistical analysis of the age distribution data from the ship's population, initially presented in Exercise 3. The goal is to gain insights into the demographic structure of the passengers and crew on board by calculating the mean and standard deviation of the age data. These statistical measures are foundational for understanding the dispersion of the age groups, which are crucial parameters for biometric system design in health-related decision support systems.

````python
# Importing the required libraries for this exercise
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import norm
````

````python
# counting of people at each age interval shown in age labels
age_list = [16, 23, 347, 428, 334, 398, 923, 1015 , 216, 11]
age_labels = ['0-9','10-19','20-29','30-39', '40-49','50-59','60-69','70-79','80-89','90-99']
````

````python
# Creating the vector of the indexes of the age list
expanded_age_vector = []

index = 0
for index, count in enumerate(age_list):
    expanded_age_vector.extend([index] * count)
expanded_age_vector = np.array(expanded_age_vector)
````

````python
# Calculate the mean and standard deviation of the vector
mean_value = np.mean(expanded_age_vector)
standard_deviation = np.std(expanded_age_vector)
print(f'Mean: {mean_value}, Standard Deviation: {standard_deviation}')
````

After calculating the mean and standard deviation, we fit a normal distribution to the expanded age data. The mean provides a measure of the central age group, while the standard deviation gives us an idea of the spread of the age groups around the mean. 


Mean: 5.33


Standard Deviation: 1.84

Generating a normal distribution curve without the smoothing of the vectorized data, we can see that the curve is not smooth and is jagged. 

````python
# Generate points for the normal distribution curve
plt.figure(figsize=(10, 6))
plt.plot(expanded_age_vector, norm.pdf(expanded_age_vector, mean_value, standard_deviation))
plt.title('Normal Distribution of Age on Diamond Princess Cruise Ship')
plt.xlabel('Age Group Intervals')
plt.ylabel('Probability Density')
plt.grid(True)
````

## Analysis of the Normal Distribution Curve
The curve should be smooth without sharp corners or flat segments, as normal distributions are continuous and differentiable everywhere this is what we accomplish in the next part of the exercise. 

Smoothing the expanded age vector the normal distribution curve is smooth and is a better representation of the data.

````python
# Smoothing the new vector
expanded_age_vector = np.arange(min(expanded_age_vector), max(expanded_age_vector), 0.01)
````

````python
# Plot the smooth normal distribution curve
plt.figure(figsize=(10, 6))
plt.plot(expanded_age_vector, norm.pdf(expanded_age_vector, mean_value, standard_deviation))
plt.title('Smooth Normal Distribution of Age on Diamond Princess Cruise Ship')
plt.xlabel('Age Group Intervals')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
````

## Analysis of the Smooth Normal Distribution Curve
The plotted normal distribution illustrates the probability density of age group intervals aboard the Diamond Princess cruise ship. The distribution appears to be symmetric around the mean, indicating that the age groups are evenly distributed on both sides of the average. The smoothness of the curve also suggests that there are no significant outliers or irregularities in the age data.  The plot indicates that the 70-79 years old age interval has the highest frequency, followed closely by 60-69 years old. This is seen by the counts of 1015 and 923, respectively, which suggests a significant concentration of the population within the older age groups. The mean value provides the balance point of the distribution and is influenced heavily by the age intervals with the most individuals. Given the shape of the distribution and the highest counts in the older age intervals, we can infer that the mean age group index is likely skewed towards the older population. The standard deviation measures the amount of variation of age intervals from the mean. The distribution curve appears to have a moderate spread, suggesting a relatively diverse but not extremely dispersed age demographic. The older age groups, which dominate this distribution, are typically more vulnerable to COVID-19. Therefore, any health and safety protocols or biometric systems designed for this population need to take into account the higher risks associated with older age groups.
